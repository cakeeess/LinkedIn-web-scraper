```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import getpass
import time

url = "https://www.linkedin.com/jobs/collections/recommended/?currentJobId=4000360078&discover=recommended&discoveryOrigin=JOBS_HOME_JYMBII"
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get(url)
email = "sssamriddhii@gmail.com"
inp_username = driver.find_element(By.ID, 'username')
inp_username.send_keys(email)

pwd = getpass.getpass("Enter your LinkedIn password: ")
inp_pwd = driver.find_element(By.ID, 'password')
inp_pwd.send_keys(pwd)

element = driver.find_element(By.XPATH, '//*[@id="organic-div"]/form/div[3]')
element.click()


li_elements = driver.find_elements(By.XPATH, "//li[contains(@class, 'artdeco-pagination__indicator')]")
xpath_list=[]
# Loop through the li elements and extract the XPath of the button element
for index, li in enumerate(li_elements):
    try:
        # Locate the button element within the current li element
        button = li.find_element(By.XPATH, ".//button")
        
        # Get the XPath of the button element
        xpath = driver.execute_script(
            "function absoluteXPath(element) {"
            "var comp, comps = [];"
            "var parent = null;"
            "var xpath = '';"
            "var getPos = function(element) {"
            "var position = 1, curNode;"
            "if (element.nodeType == Node.ATTRIBUTE_NODE) {"
            "return null;"
            "}"
            "for (curNode = element.previousSibling; curNode; curNode = curNode.previousSibling){"
            "if (curNode.nodeName == element.nodeName){"
            "++position;"
            "}"
            "}"
            "return position;"
            "};"
            "if (element instanceof Document) {"
            "return '/';"
            "}"
            "for (; element && !(element instanceof Document); element = element.nodeType == Node.ATTRIBUTE_NODE ? element.ownerElement : element.parentNode) {"
            "comp = comps[comps.length] = {};"
            "switch (element.nodeType) {"
            "case Node.TEXT_NODE:"
            "comp.name = 'text()';"
            "break;"
            "case Node.ATTRIBUTE_NODE:"
            "comp.name = '@' + element.nodeName;"
            "break;"
            "case Node.ELEMENT_NODE:"
            "comp.name = element.nodeName;"
            "break;"
            "}"
            "comp.position = getPos(element);"
            "}"
            "for (var i = comps.length - 1; i >= 0; i--) {"
            "comp = comps[i];"
            "xpath += '/' + comp.name.toLowerCase();"
            "if (comp.position !== null) {"
            "xpath += '[' + comp.position + ']';"
            "}"
            "}"
            "return xpath;"
            "}"
            "return absoluteXPath(arguments[0]);", button)

        # Print the XPath of the button element
        xpath_list.append(xpath)
        #print(f"XPath of button {index + 1}: {xpath}")
        #elem = driver.find_element(By.XPATH, xpath)
        #elem.click()
        #print(f"Done page: {index + 1}")

    
    except Exception as e:
        print(f"Could not locate button for li {index + 1}: {e}")

print(len(xpath_list))
        

job_details = []
job_titles = []
i=1
# Find job titles using the same method
jobs = driver.find_elements(By.CLASS_NAME, 'full-width.artdeco-entity-lockup__title.ember-view')
# Find job addresses using CLASS_NAME
addresses = driver.find_elements(By.CLASS_NAME, "job-card-container__metadata-item")

links = driver.find_elements(By.CLASS_NAME, "job-card-container__link")

# Extract the text for both job titles and addresses
for address,link in zip(addresses,links):
    job_details.append((address.text,link.get_attribute("href")))

for job in jobs:
    job_titles.append(job.text)
    
for i in range(len(job_titles)):
    print(f"Job {i}",job_titles[i].split('\n')[0])

# Output the job titles and addresses
for i, (address, link) in enumerate(job_details, i):
    print(f" Address {i}: {address}, Link {i}: {link}")
    
for i in range(len(job_titles)):
    print(f"Job {i}",job_titles[i].split('\n')[0])

if len(xpath_list) > 1:
    elem = driver.find_element(By.XPATH,xpath_list[1])
    time.sleep(3)
    elem.click()
    WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "job-card-container__metadata-item ")))
    
    for j in range(1,len(xpath_list)):

        # Find job titles using the same method
        jobs = driver.find_elements(By.CLASS_NAME, 'full-width.artdeco-entity-lockup__title.ember-view')
        # Find job addresses using CLASS_NAME
        addresses = driver.find_elements(By.CLASS_NAME, "job-card-container__metadata-item")
        
        links = driver.find_elements(By.CLASS_NAME, "job-card-container__link")

        # Extract the text for both job titles and addresses
        for address, link in zip(addresses,links):
            job_details.append((address.text, link.get_attribute("href")))
            
        for job in jobs:
            job_titles.append(job.text)

        # Output the job titles and addresses
        for i, (address, link) in enumerate(job_details, i):
            print(f" Address {i}: {address}, Link {i}: {link}")
            
        for i in range(len(job_titles)):
            print(f"Job {i}",job_titles[i].split('\n')[0])

        elem = driver.find_element(By.XPATH,xpath_list[j])
        time.sleep(3)
        elem.click()
        WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "job-card-container__metadata-item ")))
   

driver.quit()
```

```python
title = []
loc = []
link = []
for i in range(len(job_details)):
    loc.append(job_details[i][0])
    link.append(job_details[i][1])
for i in range(len(job_titles)):
    title.append(job_titles[i].split('\n')[0])
```
